```python
import geopy
geolocator = geopy.Nominatim(user_agent='http')
```

```python
from geopy.extra.rate_limiter import RateLimiter
reverse = RateLimiter(geolocator.reverse, min_delay_seconds=1)
```

```python
#Se carga el csv
df = pd.read_csv("zip_data_2.csv",encoding= 'unicode_escape')
```

```python
#Se remueven missing values o Na:
df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=True)
df
```

```python
df['Location'] = df.progress_apply(
lambda row: reverse((row['lat_dir'], row['lng_dir'])),
    axis=1)
```

```python
def parse_zipcode(location):
    if location and location.raw.get('address') and location.raw['address'].get('postcode'):
               return location.raw['address']['postcode']
    else:
            return None
df['Zipcode'] = df['Location'].apply(parse_zipcode)
```
